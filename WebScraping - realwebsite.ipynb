```python
from bs4 import BeautifulSoup

#import library "requests" (pip install from terminal), it requests information form a specific website
import requests

URL = 'https://www.usine-digitale.fr/annuaire-start-up/'

for pages in range(1,4):
    
    html_text = requests.get(URL + str(pages) + '/').text
    #print(html_text)

    #Create an instance from BeautifulSoup, parser lxml
    soup = BeautifulSoup(html_text,'lxml')
    startups = soup.find_all('section', class_ = 'blocType1')

    for startup in startups:
        startup_name = startup.find('h2', class_ = 'titreType2').text
        descriptions = startup.find('p', class_ = 'chapoType1 isNoMobile').text.replace('#','')
        secteur = startup.find('blockquote', class_ = 'keywordEtiquette').text.replace('\n','')
        more_info = 'https://www.usine-digitale.fr' + startup.a['href']

#    print(f'''
#    Start-up Name: {startup_name}
#    Secteur: {secteur}
#    Description: {descriptions}
    #''')

        print(f"Start-up Name: {startup_name.strip()}")
        print(f"Secteur: {secteur.strip()}")
        print(f"Description: {descriptions.strip()}")
        print(f'More Info: {more_info}')
        print('')

        sys.stdout=open("TestSaveTxt.txt","w")
        print("Hello World")
        sys.stdout.close()

```
